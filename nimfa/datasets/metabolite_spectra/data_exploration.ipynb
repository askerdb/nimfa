```python
import pickle as pkl
import numpy as np
from sklearn.decomposition import PCA,TruncatedSVD
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
X, bins, scan_names = pkl.load(open("metabolite_matrix.pkl", "rb"))
X = X.todense()
pca = TruncatedSVD(n_components=2)
pca.fit(X)
pcas = pca.transform(X)
print(pca.explained_variance_ratio_)

plotdf = pd.DataFrame(zip(pcas[:, 0], pcas[:, 1]), columns = ["PC1", "PC2"])
sns.scatterplot(data= plotdf, x = "PC1", y = "PC2")
```

```python
import umap
fit = umap.UMAP()
%time u = fit.fit_transform(X)
plotdf = pd.DataFrame(zip(u[:, 0], u[:, 1], [x.split("_")[0] for x in scan_names]), columns = ["PC1", "PC2", "Library"])
sns.scatterplot(data= plotdf, x = "PC1", y = "PC2", hue = "Library")
```
